```
import math #instruccion para llamar la biblioteca

x = 16 #declaracion de variable

#Calculemos la raiz de 16

resultado = math.sqrt(x) #funcion raiz cuadrada
print("La raiz de 16 es:", resultado)
```

```
import numpy as np

arr = np.array([1, 2, 3, 4])
#Realizar la operacio sobre el array
squared = np.sqrt(arr) #raiz de cada elemento
print("la raiz de cada elemento es:", squared)
```

```
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y)
plt.title("Grafico de f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
